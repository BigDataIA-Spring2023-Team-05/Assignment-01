```python
import boto3
import boto3.s3
```

```python
s3 = boto3.resource('s3', region_name='us-east-1')
```

```python
# s3.Bucket("noaa-goes18")
my_buc=s3.Bucket(name='noaa-goes18')
```

```python
for objects in my_buc.objects.filter(Prefix="ABI-L1b-RadF"):
    print(objects.key)
```

```python
for key in my_buc.get_all_keys(prefix='ABI-L1b-RadF/', delimiter='/'):
    print(key.name)
```

```python
import boto3 
s3 = boto3.client("s3")
response = s3.list_objects_v2(
            Bucket="noaa-goes18",
            Prefix ='ABI-L1b-RadF/',
            MaxKeys=100 )
```

```python
s3client = boto3.client('s3')
bucket = 'noaa-goes18'
startAfter = 'ABI-L1b-RadF'


theobjects = s3client.list_objects_v2(Bucket=bucket, StartAfter=startAfter )
for object in theobjects['Contents']:
    direcoryName = object['Key'].encode("string_escape").split('/')
    print(direcoryName[1])
```

```python
from boto3.session import Session

s3client = session = boto3.session.Session().client('s3')
theobjects = s3client.list_objects_v2(Bucket=bucket, StartAfter=startAfter )
for object in theobjects['Contents']:
    print(object['Key'])
```

```python
client = boto3.client('s3')
bucket = 'noaa-goes18'
folders = set()

for prefix in client.list_objects(Bucket=bucket, Delimiter='/')['CommonPrefixes']:
    folders.add(prefix['Prefix'][:-1])
    
print(folders)
```
